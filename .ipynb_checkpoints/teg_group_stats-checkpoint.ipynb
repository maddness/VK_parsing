```python
import vk
import re
import time
import datetime
import json

from dateutil.relativedelta import relativedelta
```

```python
# session = vk.AuthSession(app_id='5342290', user_login='fireroom@gmail.com', user_password='')
session = vk.Session()

api = vk.API(session)
```

```python
def get_users():
    users_all = []
    count_all = int(api.groups.getMembers(group_id='talks_ted', count=0)['count'])
    
    offset = 1000
    current_offset = 0
    
    for batch in range(int(count_all / offset) + 1):
        users_all += api.groups.getMembers(group_id='talks_ted', count=1000,
                                           offset=current_offset, fields = ['sex', 'bdate'])['users']
        current_offset += offset
    
    return users_all

users_all = get_users()
```

```python
with open('users.json', 'w') as outfile:
    json.dump(users_all, outfile)
```

```python
def count_gender(users):
    gender_map = {0: '?', 1: 'female', 2: 'male'}
    gender = {'male': 0, 'female': 0, '?': 0}

    for user in users_all:
        gender[gender_map[user['sex']]] += 1
    
    return gender
```

```python
def count_birthdays(users): 
    ages = { '<=10': 0, '11-20': 0, '21-30': 0, '?': 0, '>=31': 0 }
    today = datetime.now()
    
    for user in users:
        if 'bdate' in user:
            match = re.fullmatch(r'(\d{1,2})\.(\d{1,2})\.(\d\d\d\d)', user['bdate'])
            if match:
                birthday = date(int(match.group(3)), int(match.group(2)), int(match.group(1)))
                diff = relativedelta(today, birthday).years
                
                if diff <= 10: ages['<=10'] += 1
                if 11 <= diff <= 20: ages['11-20'] += 1
                if 21 <= diff <= 30: ages['21-30'] += 1
                if diff >= 31: ages['>=31'] += 1
                    
        else:
            ages['?'] += 1
                
    return ages
```
